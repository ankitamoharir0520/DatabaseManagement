### Conceptual Model

A conceptual model is a representation of a system, made of the composition of concepts which are used to help people know, understand, or simulate a subject the model represents. It is also a set of concepts. The term conceptual model may be used to refer to models which are formed after a conceptualization or generalization process.

Our conceptual model is based on Movies Dataset which contains of four entities that represent consumers as actors, producers as directors, companies as movie producing companies and movies as the main entity. The actor entity consists of the attributes : actor name, year the actor was awarded, award the actor won, and movies the actor has acted with actor as the primary key. The director entity consists of the attributes : director name, year the director was awarded and director's award category with director name as the primary key attribute. The company entity consists of the attributes : company name, country of company, headquarters of the company and year the company was established with company name as the primary key attribute. The movies entity consists of the following attributes : movie name, budget of the movie, movie company, country, genre of the movie, movie rating, movie score, actor name, movie votes, movie writer, year, director name with attributes movie name, company name, actor name and dirctor name as the foreign key attributes.

In our model, various fields like gross, genre, movie name,etc act as attributes.

4. How did you choose your keys? Which are unique?
    
    A primary key, also called a primary keyword, is a key in a relational database that is unique for each record. A relational database must always have one and only one primary key.
    
    A foreign key is a field (or collection of fields) in one table that uniquely identifies a row of another table or the same table. In simpler words, the foreign key is defined in a second table, but it refers to the primary key or a unique key in the first table.
    
    A Super Key is a set of one or more attributes that are taken collectively and can identify all other attributes uniquely. As our Super key. Each super key is able to uniquely identify each tuple (record).
    
    Unique Key - A Unique Key is a set of one or more fields/columns of a table that uniquely identify a record in database table. It is like Primary key but it can accept only one null value and it can not have duplicate values.
    
    Although several candidate keys may exist, one of the candidate keys is selected to be the primary key. Only one Candidate Key can be Primary Key. 

    While choosing the keys, we must consider the mapping cardinality of the relationship set as to what are the candidate keys and also we need to consider semantics of the relationship set in selecting the primary key in case of more than one candidate key.

    With the help of movies table , a relationship was established between actor, director and company of movies where these attributes in movie table act as foreign keys and primary key in their respective entities. Also, data from instagram and twitter tables are mapped to movies table as primary and foreign keys as shows in the above conceptual model diagram.

5. Did you model hierarchies using the “ISA” design element? Why or why not?

    After establishing a class hierarchy with the Entity/Relation model, the principle of generalization is used to identity the class hierarchy and the level of abstraction associated with each class.  Generalization implies a successive refinement of the class, allowing the superclasses of objects to inherit the data attributes and behaviors which apply to the lower levels of the class.  Generalization establishes "taxonomy hierarchies", which organize the classes according to their characteristics, usually in increasing levels of detail.  Generalization begins at a very general level, and proceeds to a specific level, with each sub-level having its own unique data attributes and behaviors.
    
    the ISA relationship is used to create a hierarchy within the object class, and all of the lower-level classes will inherit the behaviors.  The ISA relationship is used to model the hierarchy which is created as the class entity is decomposed into its logical sub-components.

```python
#Dataframe for twitter tweets and data based on the movie data

posts =[] #user tweet on the movie
favourites =[] #favourites count on the post
retweets = []  #number of retweets on the post
time = [] #time the post was posted
user_id = [] #user id of the user who posted
name = [] #name of the user who posted the post
followers = [] #number of followers
movie1 = [] #name of the movie for which post is posted
screen_name = [] #twitter handle name of the user who posted

movie_names=movies['movie'].tolist()
top = movie_names[:100]
for movie_name in top:
    fetched_tweets = api.search(movie_name)
    for tweet in fetched_tweets:
        posts.append(tweet.text)
        favourites.append(tweet.favorite_count)
        retweets.append(tweet.retweet_count)
        time.append(tweet.created_at)
        user_id.append(tweet.id_str)
        followers.append(tweet.user.followers_count)
        screen_name.append(tweet.user.screen_name)
        name.append(tweet.user.name)
        movie1.append(movie_name)
```

```python
#Dataframe to store twitter data

twitter_data = pd.DataFrame(data = name, columns = ['UserName'])
twitter_data['movie'] = movie1
twitter_data['Userid'] = user_id
twitter_data['TweetFollowers'] = followers
twitter_data['Tweet'] = posts
twitter_data['TweetTime'] = time
twitter_data['Retweets'] = retweets
twitter_data['ScreenName'] = screen_name
twitter_data
```

### Database tables Creation using the dataframes 

```python
#Establishing connection with sqlite using sqlalchemy

from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
#Creating movies database from movies dataframe

movies.to_sql('movies', con=engine, if_exists='append')
```

```python
#Querying on movies database

engine.execute("SELECT * FROM movies").fetchall()  
```

### Queries

a) What user posted this (e.g. tweet, facebook post, IG post, etc.)?

```python
#Converting twitter_data and Twitter_User dataframes into sql database

twitter_data.to_sql('twitter_data', con=engine, if_exists='append')
Twitter_User.to_sql('Twitter_User', con=engine, if_exists='append')
```

```python
#Retrieving a user's name for a particular post

engine.execute("SELECT distinct UserName,tweet FROM twitter_data WHERE movie = 'FLIGHT OF THE NAVIGATOR'").fetchall()
```

```python
#Retrieving username,tweet and movie name by joining both twitter tables with help of screen name therefore showing which user has posted

engine.execute("SELECT [twitter_data].UserName, [twitter_data].Tweet, [twitter_data].movie FROM twitter_data INNER JOIN Twitter_User ON [twitter_data].ScreenName = [ ].ScreenName").fetchall() 
```

```python
#Retrieving what a particular user has posted

engine.execute("SELECT DISTINCT Tweet FROM twitter_data WHERE movie = 'ALIENS'").fetchall()
```

b) When did the user post this (e.g. tweet, facebook post, IG post, etc.)??

```python
#Retrieving the time the user has posted the tweet

engine.execute("SELECT min(tweettime) FROM twitter_data where tweet like '%Labour members know that if they turn their back on Corbyn%' ").fetchall()
```

c)What posts has this user posted in the past 24 hours?

```python
#Retrieving posts of last 24 hours

from datetime import datetime, timedelta
end = "'"+str(datetime.now() - timedelta(days = 1))+"'"

print(end) 

engine.execute("SELECT UserTweets,DateOfTweets FROM Twitter_User where DateOfTweets >= "+end+ " ORDER BY DateOfTweets").fetchall()
```

d) How many post has this user posted in the past 24 hours?

e) What keywords/ hashtags are popular?
