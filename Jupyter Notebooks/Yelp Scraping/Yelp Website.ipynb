```python
from requests import get
```

```python
#URL to be scraped; here 'Yelp' to retreive restaurants from Boston

url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=Boston%2C+MA+02115&ns=0"
```

```python
#if response = 200, it means that the website allows user to scrape the data

response = get(url)
response
```

```python
#BeautifulSoup - package for parsing HTML and XML documents

from bs4 import BeautifulSoup as bs ## importing Beautiful Soup
html_soup = bs(response.text,'html.parser')## python's in built library HTML parser
type(html_soup)
```

```python
#Below are the columns in the restaurant dataset:
#•	restaurant_names #gives the name of restaurants
#•	restaurant_contacts #displays the contact of restaurants 
#•	restaurant_addresses #shows the addresses of restaurants
#•	restaurant_areas #area of restaurants are displayed here
#•	restaurant_reviews #gives the number of people who have given reviews
#•	restaurant_prices #Prices of restaurants displayed
#•	restaurant_cuisines #types of cuisines are shown here
#•	restaurant_distances #shows the distance of the restaurant
#•	restaurant_feedbacks #Shows customer feedbacks on restarants
#•	restaurant_deliveries #shows if restaurant delivers food 

#list to store scraped value data in:

restaurant_names = []
restaurant_contacts = []
restaurant_addresses = []
restaurant_areas = []
restaurant_reviews = []
restaurant_prices = []
restaurant_cuisines = []
restaurant_distances = []
restaurant_feedbacks = []

#extract data from individual movie container
for i in range(0,20): #loop for 10 pages of data
    url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=Boston%2C%20MA%2002115&ns="+str(i)+"&start="+str((i-1)*30)
    response = get(url)
    
    #python's in built library HTML parser for whole package
    html_soup = bs(response.text,'html.parser')
    
    all_types=html_soup.select('.u-padding-b3.border--top__373c0__19Owr') #data of 10 restaurants/page with total 10 pages 
    
    for q in range(1,len(all_types)):
        stringtext=str(all_types[q]) #data of one form with 10 restaurant information
        if(len(stringtext))>10:
            
            #python's in built library HTML parser for scraping current webpage data
            html_soup1=bs(stringtext,'html.parser') #python's in built library HTML parser for scraping current webpage data 
            
            restaurant_name = html_soup1.select(".alternate__373c0__1uacp .link-size--inherit__373c0__2JXk5")
            
            restaurant_contact = html_soup1.select(".text-align--right__373c0__3fmmn > .display--inline-block__373c0__2de_K")
            
            restaurant_address = html_soup1.select(".display--inline-block__373c0__2de_K.u-space-t1 .domtags--span__373c0__1VGzF")
            
            restaurant_area = html_soup1.select(".text-align--right__373c0__3fmmn .border-color--default__373c0__2oFDT .border-color--default__373c0__2oFDT")
            
            restaurant_review = html_soup1.select(".text-color--mid__373c0__3G312")
            
            restaurant_price = html_soup1.select(".text-bullet--after__373c0__1ZHaA")
            
            restaurant_cuisine = html_soup1.select(".link-color--inherit__373c0__15ymx")
            
            restaurant_feedback = html_soup1.select(".alternateStyling__373c0__2ithU")

            #extracting data from website and saving it as a string
            names_only = ''.join([p.get_text() for p in restaurant_name])
            restaurant_names.append(names_only)
            
            contact_only = ''.join([p.get_text() for p in restaurant_contact])
            restaurant_contacts.append(contact_only)
            
            address_only = ''.join([p.get_text() for p in restaurant_address])
            restaurant_addresses.append(address_only)
            
            area_only = ''.join([p.get_text() for p in restaurant_area])
            restaurant_areas.append(area_only)
        
            reviews_only = ''.join([p.get_text() for p in restaurant_review])
            restaurant_reviews.append(reviews_only)
           
            price_only = ''.join([p.get_text() for p in restaurant_price])
            restaurant_prices.append(price_only)
            
            cuisine_only = ''.join([p.get_text() for p in restaurant_cuisine])
            restaurant_cuisines.append(cuisine_only)
            
            feedback_only = ''.join([p.get_text() for p in restaurant_feedback])
            restaurant_feedbacks.append(feedback_only)
```

```python
import pandas as pd
from pandas import DataFrame
dataframe1= pd.DataFrame();
dataframe1['business_names'] = restaurant_names 
dataframe1['Address'] = restaurant_addresses
dataframe1['Contact Number'] = restaurant_contacts
dataframe1['Area'] = restaurant_areas
dataframe1['Reviews'] = restaurant_reviews
dataframe1['Price Information'] = restaurant_prices
dataframe1['category'] = restaurant_cuisines
dataframe1['Feedback Reviews'] = restaurant_feedbacks
```

```python
dataframe1.to_csv(index=False)
```
